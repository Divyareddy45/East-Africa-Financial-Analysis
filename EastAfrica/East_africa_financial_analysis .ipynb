```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv("financial_inclusion_dataset.csv")
```

```python
df
```

```python
df['age'].max()
```

```python
df.loc[(df['age'] <= 0) | (df['age'] >= 150), 'age'] = np.nan
```

```python
df['age'].max()
```

```python
#Normalize has_bank_account field (yes, no, unknown → 1/0/NaN).
```

```python
df['has_bank_account'].unique()
```

```python
df['has_bank_account'] = df['has_bank_account'].astype(str).str.lower().str.strip()
mapping = {'yes': 1, '1': 1,'no': 0, '0': 0}
df['has_bank_account']=df['has_bank_account'].map(mapping)
```

```python
df['has_bank_account'].unique()
```

```python
#Remove or impute rows with negative income.
```

```python
df.loc[(df['monthly_income']<0, 'monthly_income')]=np.nan
```

```python
df.isnull().sum()
```

```python
df['age']=df['age'].fillna(df['age'].median())
```

```python
num_col=df.select_dtypes(include=['number']).columns
num_col

for col in num_col:
    median_val=df[col].median()
    df[col] = df[col].fillna(median_val)
```

```python
cat_col=df.select_dtypes(exclude=['number']).columns
cat_col

for col in cat_col:
    mode_val=df[col].mode()[0]
    df[col]=df[col].fillna(mode_val)
```

```python
df.isnull().sum()
```

```python
df=df.drop('monyhly_income', axis=1)
```
